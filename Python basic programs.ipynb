2. Create list of elements and slice and dice it

```python
List = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print("\nOriginal List:\n", List)
print("\nSliced Lists: ")
print(List[3:9:2])
print(List[::2])
print(List[::])
```

```python
import random
def dice():
    n = input( "Enter # of dice [3,6]: " )
    n = int( n )
    diceList = []
    for row in range(0, n):
        diceList.append(random.randint(1, 6))
    print( "\nYou have rolled: ", diceList)
dice()
```

```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = 'https://www.yelp.com/biz/milk-and-cream-cereal-bar-new-york?osq=Ice+Cream'
ourUrl = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(ourUrl,'html.parser')
```

```python
print(soup.prettify())
```

```python
review = []
for i in soup.find_all('div',{'class':'review-content'}):
  per_review=i.find('p')
  print(per_review)
  review.append(per_review)
```
